##Задание для гениев *
Агрегируйте числовой признак по категориальному признаку с использованием метода **без использования** библиотеки pandas, и вычислите медиану и среднее значение для каждой категории.

```python
#Агрегация
import numpy

def np_mean(x):
    return sum(x) / len(x)

def np_median(x):
    l = len(x)
    i = l // 2
    if l % 2:
        return sorted(x)[i]
    return sum(sorted(x)[i -1 : i + 1]) / 2

df2 = df.to_numpy().transpose()

print('Медиана:\n\t'
      f'Age:          {np_median(df2[0])}\n\t'
      f'RestingBP:    {np_median(df2[3])}\n\t'
      f'Cholesterol:  {np_median(df2[4])}\n\t'
      f'FastingBS:    {np_median(df2[5])}\n\t'
      f'MaxHR:        {np_median(df2[7])}\n\t'
      f'Oldpeak:      {np_median(df2[9])}\n\t'
      f'HeartDisease: {np_median(df2[11])}\n\t'
      )

print('Среднее значение:\n\t'
      f'Age:          {np_mean(df2[0]):.5f}\n\t'
      f'RestingBP:    {np_mean(df2[3]):.5f}\n\t'
      f'Cholesterol:  {np_mean(df2[4]):.5f}\n\t'
      f'FastingBS:    {np_mean(df2[5]):.5f}\n\t'
      f'MaxHR:        {np_mean(df2[7]):.5f}\n\t'
      f'Oldpeak:      {np_mean(df2[9]):.5f}\n\t'
      f'HeartDisease: {np_mean(df2[11]):.5f}\n\t'
      )
```
